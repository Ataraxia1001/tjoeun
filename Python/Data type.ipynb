```python
num1 = 100
num2 = 20

# and 조건
log_result = num1 >= 50  and num2 <= 10 # 참 and 거짓
print(log_result)
log_result2 = num1 <= 50 and num2 >= 10 # 거짓 and 참
print(log_result2)
log_result3 = num1 >= 50 and num2 >= 10 # 참 and 참
print(log_result3)
log_result3 = num1 <= 50 and num2 <= 10 # 거짓 and 거짓
print(log_result3)
```

```python
# or 조건
log_result5 = num1 >= 50  or num2 <= 10 # 참 or 거짓
print(log_result5)
log_result6 = num1 <= 50 or num2 >= 10 # 거짓 or 참
print(log_result6)
log_result7 = num1 >= 50 or num2 >= 10 # 참 or 참
print(log_result7)
log_result8 = num1 <= 50 or num2 <= 10 # 거짓 or 거짓
print(log_result8)
```

```python
log_result9 = not(num1 >= 50) # not true
print(log_result9)
```

```python
'''
%양식문자
%s : 문자열(string)
%c : 한개문자
%d : 정수
%f : 실수
%o : 8진수
%x : 16진수
%% : %출력
'''
```

```python
print('원주율 = ', 3.14159)
```

```python
print('원주율 = ', format(3.14159, "8.3f")) #소수점 3째자리까지 실수로 표현, 앞자리 8개 
```

```python
print('금액 = ', format(1000000000000, "9,d")) # 3자리당 콤마 한개씩
```
