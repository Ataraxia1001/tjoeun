# 시도별 전출입 인구수 데이터 시각화

```python
df = pd.read_excel('시도별 전출입 인구수.xlsx', engine = 'openpyxl', header=0)
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.fillna(method='ffill') # ffill: forward fill
# ffill: propagate last valid observation forward to next valid backfill
# bfill: use next valid observation to fill gap
```

```python
# NaN: Not a number: As the name implies, it is used to denote that the value of an object is not a number. 
# There are many ways that you can generate this error, one being invalid math opertaions such as 0/0 or sqrt(-1)

# undefined: It means that the object doesn't have any value, therefore undefined. 
# This occurs when you create a variable and don't assign a value to it.

# null: It means that the object is empty and isn't pointing to any memory address.
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'], axis=1) # 전출지별이 모두 서울이라 그냥 제거
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace = True)
df_seoul.set_index('전입지', inplace = True)
```

```python
df_seoul
```

```python
# 서울에서 경기도로 이동한 인구 데이터 값만 선택
sr_one = df_seoul.loc['경기도']
print(type(sr_one))
```

```python
plt.plot(sr_one.index, sr_one.values)   # Series : index + value
```

```python
plt.figure(figsize=(12,6)) # 그림사이즈(가로인치, 세로인치)
plt.plot(sr_one.index, sr_one.values, marker = 'o', markersize=7)
plt.title('서울 -> 경기 인구 이동') # 차트의 제목을 추가
plt.xlabel('기간', size = 20) # axis name
plt.xticks(size = 10, rotation = 45) # rotation: angle
plt.ylabel('이동인구수', size = 20)
plt.ylim(50000, 700000)  # give range of y axis (min, max)
plt.legend(fontsize=15, labels=['서울->경기'], loc='best')
#주석표시
plt.annotate("인구이동 증가(1970-1995)",    # 텍스트 입력
             xy=(10, 350000),              # 텍스트 위치 기준점
             rotation=25,                  # 텍스트 회전각도
             va='baseline',                # 텍스트 상하정렬
             ha='center',                  # 텍스트의 좌우정렬
             fontsize=15                   
             )
plt.annotate("인구이동 감소(1995-2017)",    # 텍스트 입력
             xy=(40, 480000),              # 텍스트 위치 기준점
             rotation=-18,                  # 텍스트 회전각도
             va='baseline',                # 텍스트 상하정렬
             ha='center',                  # 텍스트의 좌우정렬
             fontsize=15                   
             )

plt.show() # 변경사항 저장하고 그래프로 출력
```

```python
df
```

```python
df = df.T
df.tail(2)
```

```python
# 증감율(변동률) 계산
df = df.rename(columns={'합계':'총발전량'})
df
```

```python
df['총발전량 - 1년'] = df['총발전량'].shift(1)
df.head(4)
```

```python
df['증감율'] = ((df['총발전량'] / df['총발전량 - 1년']) -1 ) * 100
df.head(4)
```

```python
# 2축 그래프 그리기
ax1 = df[['수력', '화력']].plot(kind = 'bar', figsize=(20, 10), width = 0.7, stacked = True)
ax2 = ax1.twinx() # two y axis on one x axis
ax2.plot(df.index, df.증감율, ls='--', marker = 'o', markersize = 20, color = 'green', 
         label = '전년대비 증감율(%)')  
#‘line’ : line plot (default)
# ls = line style

ax1.set_ylim(0, 500)
ax2.set_ylim(-50, 50)

ax1.set_xlabel('연도', size = 20)
ax1.set_ylabel('발전량(억 KwH)')
ax2.set_ylabel('전년대비 증감율')

plt.title('북한 전력 발전량(1990~2016', size = 25)
ax1.legend(loc='upper left')
plt.show()
```
