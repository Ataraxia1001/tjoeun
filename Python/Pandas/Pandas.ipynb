```python
student1 = pd.Series({'국어':100, '영어':80, '수학':90})
student2 = pd.Series({'수학':80, '국어':90, '영어':90})
# 두학생의 과목별 점수 사칙연산
add1 = student1 + student2 # 덧셈
sub = student1 - student2  # 뺄셈
mul = student1 * student2  # 곱셈
div = student1 / student2  # 나눗셈
print(add1, sub, mul, div)
```

```python
# 사칙연산 결과를 데이터프레임으로 합치기 (series, --> 데이터프레임)
result = pd.DataFrame([add1, sub, mul, div], index = ['덧셈', '뺄셈', '곱셈', '나눗셈'])
print(result)
```

```python
# 결측치가 있는경우 사칙연산
student3 = pd.Series({'국어':np.nan, '영어':80, '수학':90})
student4 = pd.Series({'수학':80, '국어':90, '영어':90})
# 두학생의 과목별 점수 사칙연산
add2 = student3 + student4 # 덧셈
sub2 = student3 - student4  # 뺄셈
mul2 = student3 * student4  # 곱셈
div2 = student3 / student4  # 나눗셈
print(add1, sub, mul, div)
```

```python
result2 = pd.DataFrame([add2, sub2, mul2, div2], index = ['덧셈', '뺄셈', '곱셈', '나눗셈'])
print(result2)
```

```python
# 두학생의 과목별점수 사칙연산(연산메소드 사용)
add3 = student3.add(student4, fill_value=0)
sub3 = student3.sub(student4, fill_value=0)
mul3 = student3.mul(student4, fill_value=0)
div3 = student3.div(student4, fill_value=0)
print(add3, sub3, mul3, div3)
```
