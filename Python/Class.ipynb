```python
class 세면도구:

  def __init__(self, 치약, 샴푸, 비누):
    self.치약 = 치약
    self.샴푸 = 샴푸
    self.비누 = 비누


  def 내용물출력(self, 세트이름):
    print(세트이름)
    print('치약:', self.치약):
    print('샴푸:', self.샴푸):
    print('비누:', self.비누):

class 특별세트(세면도구):

  def __init__(self, 치약, 샴푸, 비누, 면도기):
    super().__init__(치약, 샴푸, 비누)
    self.면도기 = 면도기


  def 내용물출력(self, 세트이름):
    super().내용물출력(세트이름)
    print('면도기:', self.면도기)


특01호 = 특별세트(1,2,3,1)
특01호.내용물출력('명절 특별세트 01호')

```

*메소드(기능, def) : drive, trun_off, trun_on 

```python
class 자동차():
  브랜드 = ''
  model = ''
  color = ''
  def drive(self):
     print('drive')

  def trun_off(self):
     print('trunc off')
  def trun_on(self):
     print('trunc on')
```

```python
# solution
 
class Car():
  def __init__(self, b, m, c):
    self.brand = b
    self.model = m
    self.color = c
    print(self.brand, self.model, self.color, '출고') 

  def turn_on(self):
    print(self.brand, '시동을 겁니다.')
  def turn_off(self):
    print(self.brand, '시동을 끕니다.')
  def drive(self):
    print(self.brand, '주행중입니다.')
```

```python
# solution
class Character():
  def __init__(self):
    self.health = 200
  def move(self):
    self.health -= 10
  def rest(self):
    self.health += 10
  def checkHealth(self):
    print('현재 캐릭터의 체력은', self.health)  
```

```python
class Knight(Character):
  def __init__(self):
    super().__init__()  # health parent로부터  상속
  def move(self):
    super().move()  # 기본 move를 parent로부터 상속
    self.health -= 5 # Knight는 move시 -5 추가
  def Attack(self):
    print('공격합니다.')
  

class Healer(Character):
   def __init__(self):
     super().__init__()
     self.mana = 100
   def heal(self, Character):
     Character.rest()
     self.mana -= 10
   def checkMana(self):
     print('현재 캐릭터의 마나는', self.mana, '입니다.')
```
