```python
# 숫자를 입력받아서 10의 배수면 프로그램 종료가 되도록 while 문으로 만들기

num = int(input('숫자를 입력하세요'))

while num % 10 != 0:
  print('다시 입력하세요')
  num = int(input('다시 입력하세요'))
  continue
  if num % 10 == 0:
   break
print('종료')
```

```python
# Solution

numData = []

while True:
  num = int(input('숫자를 입력: '))
  if num % 10 == 0:
    print('프로그램 종료')
    break
  else:
    print(num)
    numData.append(num)

print('내가 입력한 숫자', numData)
```

```python
# id = 'id123', pwd = 'pwd123'
# while문으로 아이디와 패스워드를 입력받아서 id가 틀리면  id를 확인하세요
# pwd 입력받아서 틀리면 pwd를 확인하세요

id_sec = 'id123'
pwd_sec = 'pwd123'


while True:
  id = input('아이디')
  pwd = input('패스워드')
  if id == id_sec and pwd == pwd_sec:
    print('로그인 하셨습니다.')
    break
  elif id != id_sec and pwd != pwd_sec:
    print('아이디를 확인하세요')
    print('패스워드를 확인하세요')
  elif id != id_sec and pwd == pwd_sec:
    print('아이디를 확인하세요')
  elif id == id_sec and pwd != pwd_sec:
    print('패스워드를 확인하세요')
  else:
    continue

# 솔루션과 비교
```

```python
# 1~100 사이에서
# 5의 배수이면서(and) 3의 배수가 아닌 합 계산하기

i = tot = 0
dataset = []
while i < 100:
  i += 1
  if i % 5 == 0 and i % 3 != 0:
    dataset.append(i)
    print(i, end = '')
    tot += i


print(f'데이터셋:{dataset}')
print(f'합은 :{tot}')
```
