```python
from statistics import mean, variance
from math import sqrt
# 산술평균
dataset = [2, 4, 5, 6, 1, 8]
def Avg(data):
  avg = mean(data)
  return avg
```

```python
print('산술평균 = ', Avg(dataset))
```

```python
# 분산/표준편차
def var_sd(data):
  avg = Avg(data)
  diff = [(d - avg)**2 for d in data] 
  var = sum(diff) / (len(data)-1)
  sd = sqrt(var)
  return var, sd
```

```python
v, s = var_sd(dataset)
print('분산', v)
print('표준편차', s)
```

```python
import random
#동전 앞면과 뒷면 난수 확률분포 함수 정의

def coin(n):
  result = []
  for i in range(n):
    r = random.randint(0, 1)
    if (r == 1):
      result.append(1)
    else:
      result.append(0)
  return result
```

```python
# MC simulation
def montaCoin(n):
  cnt = 0
  for i in range(n):
    cnt += coin(1)[0]
  result = cnt / n
  return result 
```

```python
print(montaCoin(10))
print(montaCoin(30))
print(montaCoin(10000))
print(montaCoin(10000000))
```

```python
# 2. print_5xn('string')입력받아 한줄에 다섯글자씩만 출력하는 함수를 작성
# '아이엠어보이유아어걸' --> 아이엠어보
                         #  이유알어걸
```

```python
# solution
def print_5xn(line):
  chek_num = int(len(line) / 5)  # the number of cuts
  for x in range(chek_num+1):  # loop for each cut, upto the number of cuts
    print(line[x*5:x*5 + 5]) # slice string
```

```python
print_5xn('아이엠어보이유알어걸')
```
