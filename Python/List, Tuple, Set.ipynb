리스트(튜플) 통계값 구하기
* 항목수 : len(리스트명)
* 합계 : sum(리스트명)
* 최소값 : min(리스트명)
* 최대값 : max(리스트명)
* 항목의개수: 리스트명.count(항목)


```python
lst =[1,2,3,4,5,3,4,5]

print(len(lst))
print(sum(lst))
print(min(lst))
print(max(lst))
print(lst.count(3))
print(sum(lst)/len(lst)) # mean
```

```python
tp = (1,2,3,4,5,3,4,5)
print(len(tp))
print(sum(tp))
print(min(tp))
print(max(tp))
print(tp.count(3))
print(sum(tp)/len(tp)) # mean
```

# 리스트내포(list comprehension)

```python
a = [1,2,3,4]

result =[]
for i in a:
  result.append(i*3)

result
```

```python
result2 = []
result2 = [i*3 for i in a]
result2
```

```python
a = [1,2,3,4]

# 3을 곱하고 짝수만 result3에 넣는 코드

result3 = []

for i in a:
  if i % 2 == 0:
    i=i*3
    result3.append(i)

print(result3)    
```

```python
result4 = [i*3 for i in a if i % 2 ==0]
result4
```

```python
# 구구단 코드 [2, 4, 6, 8 ... 3, 6, , ... 81, ...]
num = [1,2,3,4,5,6,7,8,9]
result5 = [i*j for i in num for j in num]
result5
```

```python
result99 = [x*y for x in range(2, 10) for y in range(1, 10)]
result99
```
