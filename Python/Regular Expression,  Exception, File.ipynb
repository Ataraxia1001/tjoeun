```python
import re
```

```python
# . 은 한개의 임의 문자임
r = re.compile("a.c")
r.search('kkk')
```

```python
r.search('abc')
```

```python
# ?는 ? 앞에 문자가 1개 존재할 수도 있고 존재하지 않을수 있는 경우
r = re.compile("ab?c") # b가 있거나 없거나
r.search('abbc') # search failed. b가 두개
```

```python
r.search('abc')
```

```python
r.search('ac')
```

```python
# *는 바로앞의 문자가 0개 이상일경우
r = re.compile('ab*c')
r.search('ac')
```

```python
r.search('abbbbc')
```

```python
r.search('a') # search failed. there is no c.
```

```python
# +는 앞의 문자가 최소 1개이상
r = re.compile('ab+c')
r.search('ac')
```

```python
r.search('abbbc')
```

```python
# ^는 시작되는 문자열
r = re.compile('^ab')
r.search('zab')
```

```python
r.search('abz')
```

```python
# {숫자}, 해당문자를 숫자만큼 반복
r = re.compile('ab{2}c') # b가 두번
```

```python
r.search('abbc')
```

```python
r.search('abbbc')
```

```python
# {숫자1, 숫자2} 행문자를 숫자1 이상 숫자2 이하 만큼 반복
r = re.compile('ab{2,8}c')
r.search('abbbbbc')
```

```python
# [ ] 안에 문자를 넣으면 문자들 중에서 한개의 문자와
# 매치하라는 의미
r = re.compile('[abc]') # [a, b, c, ... z] [a-z]
# [a-c] = [a, b, c]
r.search('z')   # 한개도 매치 안됨
```

```python
r.search('c')
```

```python
r.search('aaaaa')
```

```python
r.search('baac')
```

```python
# 알파벳 소문자
r = re.compile("[a-z]")
r.search('aBC')
```

```python
r = re.compile("[A-Za-z]")
r.search('aBC')
```

```python
r.search('1B3')
```

```python
# [^문자] ^뒤에 붙은 문자들은 제외한 모든 문자를 매치하라는 역
r = re.compile("[^abc]")
```

```python
r.search('a') # a 있어서 매치안됨
```

```python
r.search('ab') # a 있어서 매치안됨
```

```python
r.search('d')  # a,b,c 없어서 매치
```

```python
a = [1,2,3]
try: 
  x = int(input())
  y = 100 / x
  print(y, a[x])
except ValueError:
  print('입력값이 잘못됨')
except ZeroDivisionError:
  print('0으로 나눌수 없음')
except IndexError:
  print('Index 에러 발생')
```

```python
a = [1,2,3]
try: 
  x = int(input())
  y = 100 / x
  print(y, a[x])
except Exception as e:  # all errors included
  print(e)
```

```python
# else - finally
try:
  x = int(input())
  y = 100 / x
except Exception as e:  # 예외(에러) 나면 실행
  print(e)
else:           # 예외(에러) 안나면 실행
  print(y)      
finally:        # 예외발생여부와 상관없이 항상 마지막에 실행
  print('코드실행종료')
```

```python
f = open('새파일.txt', 'r')
line = f.readline()
print(line)
```

```python
f = open('새파일.txt', 'r')
while True:
  line = f.readline()
  if not line:
    break
  print(line)
f.close() 
```

```python
f = open('새파일.txt', 'a')
for i in range(11,20):
  data = "%d번째 줄입니다.\n"%i
  f.write(data)
```

```python
# 파이썬 객체 저장/로드
import pickle
person = {'name':'james', 'age':25, 'addr':'신촌'}
with open('person.pkl', 'wb') as file:
  pickle.dump(person, file) # save person(dictonary) in file
```

```python
with open('person.pkl', 'rb') as file:
  james = pickle.load(file)
print(james)
```
