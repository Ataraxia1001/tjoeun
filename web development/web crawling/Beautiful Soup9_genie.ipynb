```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import quote
```

```python
headers = {
   'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'
}
```

```python
url = 'http://mw.genie.co.kr/chart'
```

```python
response = requests.get(url, headers = headers)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
musicInfoList = soup.find('ul', attrs= {'class', 'list-music list-chart'}).find_all('li')
musicInfoList
```

```python
len(musicInfoList)
```

```python
rank = musicInfoList[0].find('div', attrs= {'class', 'ranking text-center'}).find('em').get_text()
rank
```

```python
song_name = musicInfoList[0].find('p', attrs= {'class', 'track ellipsis'}).get_text().strip()
song_name
```

```python
singer = musicInfoList[0].find('span', attrs= {'class', 'artist text-gray ellipsis'}).get_text()
singer 
```

```python
line = []
for i in range(len(musicInfoList)):
    rank = musicInfoList[i].find('div', attrs= {'class', 'ranking text-center'}).find('em').get_text()
    song_name = musicInfoList[i].find('p', attrs= {'class', 'track ellipsis'}).get_text().strip()
    singer = musicInfoList[i].find('span', attrs= {'class', 'artist text-gray ellipsis'}).get_text()
    line.append({'rank': rank, 'song':song_name, 'singer':singer})
```

```python
df = pd.DataFrame(line)
df
```
