```python
url_main = 'https://www.hanbit.co.kr/academy/books/new_book_list.html'
response_main = requests.get(url_main)
```

```python
soup_main = BeautifulSoup(response_main.text, 'html.parser')
soup_main
```

```python
href = soup_main.find('p', attrs = {'class', 'book_tit'}).find('a')['href']
href
```

   # Subpage
  

```python
url = site + href
response = requests.get(url)
sub_soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup_main
```

```python
lis = soup_main.find_all('li', attrs = {'class', 'sub_book_list'})
lis
```

```python
len(lis)
```

```python
lines = []
site='https://www.hanbit.co.kr'
for i in tqdm(range(len(lis))):
    href = soup_main.find_all('li', attrs = {'class', 'sub_book_list'})[i].find('p', attrs = {'class', 'book_tit'}).find('a')['href']
    url = site + href
    res = requests.get(url)
    sub_soup = BeautifulSoup(res.text, 'html.parser')
    title = sub_soup.find('div', attrs = {'class', 'store_product_info_box'}).find('h3').get_text()
    author = sub_soup.find('ul', attrs = {'class', 'info_list'}).find_all('li')[0].find('span').get_text().strip()
    release = sub_soup.find('ul', attrs = {'class', 'info_list'}).find_all('li')[1].find('span').get_text()
    pages = sub_soup.find('ul', attrs = {'class', 'info_list'}).find_all('li')[2].find('span').get_text().replace(' 쪽', '')
    price = sub_soup.find('span', attrs = {'class', 'pbr'}).get_text()
    lines.append([title, author, release, pages, price])
```

```python
df = pd.DataFrame(lines, 
    columns=['title', 'author', 'release', 'pages', 'price'])
df
```
